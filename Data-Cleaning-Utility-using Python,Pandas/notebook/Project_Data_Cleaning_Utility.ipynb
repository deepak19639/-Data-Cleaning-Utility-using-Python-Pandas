 **Step 3: Handle nulls**

```python
df.isnull().sum()
```

**Step 3.1 : Remove Null**

```python
df['City'] = df['City'].fillna('Unknown')
```

```python
df['Purchase Amount'] = df['Purchase Amount'].fillna(df['Purchase Amount'].median())
df['Rating'] = df['Rating'].fillna(df['Rating'].median)
```

```python
df.isnull().sum()
```

**Step 4: Handle duplicates**

```python
df.duplicated().sum()
```

**Step 4.1 : Remove Dubligate**

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

**Step 5 : Standardize Column Values**

```python
df.head()
```

 Step 3: Standardize Column Values
( casing  , whitespaces ,  formatting  )

```python
df['City'] = df['City'].str.lower().str.strip()
df['Product Category'] = df['Product Category'].str.lower().str.strip()
df['Payment Method'] = df['Payment Method'].str.lower().str.strip()

df.head()
```

2. Date Format standardize :

```python
df['Purchase Date'] = pd.to_datetime(df['Purchase Date'])
df.head()
```

**Step 6 : Standrise columns**

```python
df.columns = df.columns.str.lower().str.strip().str.replace(' ','_')

df.head()
```
